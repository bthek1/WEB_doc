## Github Connection

create ssh on computer

```sh
ssh-keygen -t ed25519 -C "your_email@example.com"
```

Add it to ssh agent

```sh
ssh-add /c/Users/YOU/.ssh/id_ed25519
```

Connect it to github
```sh
ssh -T git@github.com
```
