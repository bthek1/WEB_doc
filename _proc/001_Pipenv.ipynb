## Installation

```sh
pip install pipenv
```

## Create a env

```sh
pipenv install django
```

## Remove an env

```sh
pipenv --rm
```

## Activate already created pipenv environment

```sh
pipenv shell
```

## To Deactivate

```sh
deactivate
```

## Install/Uninstall dependencies

```sh
pipenv install django
pipenv uninstall django
```

```sh
pipenv install nose --dev
```

```sh
pipenv install -r requirements.txt
```

```sh
pipenv check
```
